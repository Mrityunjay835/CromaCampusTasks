```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
tips = sns.load_dataset("tips")
```

```python
tips.head(3)
```

```python
#a) Histogram of total_bill
plt.figure(figsize=(8,5))
sns.histplot(tips["total_bill"], bins=20, kde=True, color="skyblue")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# b) Boxplot of total_bill by day
plt.figure(figsize=(8,5))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Total Bill by Day")
plt.show()
```

```python
# Load dataset
iris = sns.load_dataset("iris")
```

```python
iris.head(3)
```

```python
# a) Scatter plot: sepal_length vs sepal_width with species as hue
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris, palette="Set1", s=100)
plt.title("Sepal Length vs Width by Species")
plt.show()
```

```python
# b) Pairplot for all numeric columns
sns.pairplot(iris, hue="species", palette="Set2")
plt.show()
```

***Using flights dataset:***
    a. Create a heatmap showing passengers by month and year.
    b. Add annotations to show exact values

```python
# Load dataset
df_flights = sns.load_dataset("flights")
```

```python
df_flights.head()
```

```python
# Pivot table to have months as columns, years as rows
flights_pivot = df_flights.pivot(index="year", columns="month", values="passengers")
```

```python
# Heatmap with annotations
plt.figure(figsize=(12,6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passengers by Month and Year")
plt.show()
```
