#Utility method

```python
def clean_str_list_column(df, col_name):
    """
    Cleans a column that contains stringified Python lists.
    Converts them into actual Python lists.
    """
    df[col_name + "_list"] = df[col_name].dropna().apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) else []
    )
    return df


def convert_votes(vote_str):
    """
    Converts votes string like '683K', '1.2M' into integer.
    Handles missing values.
    """
    if pd.isna(vote_str):
        return None
    vote_str = str(vote_str).replace(",", "").strip()
    if vote_str.endswith("K"):
        return int(float(vote_str[:-1]) * 1_000)
    elif vote_str.endswith("M"):
        return int(float(vote_str[:-1]) * 1_000_000)
    else:
        return int(float(vote_str))  # already a number


def convert_duration(duration_str):
    """
    Converts duration from '2h 4m' into total minutes as integer.
    Handles missing values.
    """
    if pd.isna(duration_str):
        return None
    duration_str = str(duration_str).lower()
    hours, minutes = 0, 0
    if "h" in duration_str:
        hours = int(duration_str.split("h")[0].strip())
        duration_str = duration_str.split("h")[1]
    if "m" in duration_str:
        minutes = int(duration_str.replace("m", "").strip())
    return hours * 60 + minutes

```

# Distribution Comparison Plan
For each feature (budget, grossWorldWide, etc.):

***Split the dataset***

   1.Group A = movies with values
   2.Group B = movies without values

***Compare distributions***
   1.Histograms / KDE plots
   2.Boxplots
   3.Summary stats (mean, median, std, skewness)

***Interpretation***
   1.If Group A & B are similar â†’ Missing may be random â†’ we can impute.
   2.If Group B looks biased (e.g., only old/low-budget films are missing) â†’ Missing is systematic â†’ we must be careful.

```python
# Split into groups
with_budget = df[df['budget'].notnull()]
without_budget = df[df['budget'].isnull()]

# Summary statistics
print("With Budget:", with_budget['budget'].describe())
print("Without Budget (Year Dist):", without_budget['Year'].describe())

# Histogram comparison
plt.figure(figsize=(12,6))
sns.histplot(with_budget['budget'], bins=50, color="blue", label="With Budget", kde=True)
sns.histplot(without_budget['Year'], bins=50, color="red", label="Missing Budget (by Year)", kde=True)
plt.legend()
plt.title("Budget Availability vs Distribution")
plt.show()
```

# What your graph confirms
1.Year-based median filling is the best choice./n
2.For missing budget, we should use the median of the same year./n
3.If a year has too few data points, we can smooth with decade-based median./n

```python
# Impute function
def impute_budget(row):
    if pd.isna(row["budget"]):
        year = row["Year"]
        if year in yearly_budget_median.index and not pd.isna(yearly_budget_median[year]):
            return yearly_budget_median[year]
        else:
            # fallback: decade median
            decade = (year // 10) * 10
            return df[df["Year"].between(decade, decade+9)]["budget"].median()
    return row["budget"]

df["budget_clean"] = df.apply(impute_budget, axis=1)
```

Plan to Compare Methods
Split the data with known budgets into train & test sets.
  * Pretend the test set budgets are "missing".
  * Impute them using both methods.
Compare imputed values with true budgets using error metrics like:
  * MAE (Mean Absolute Error)
  * RMSE (Root Mean Squared Error)

the method with lower error (or maybe a hybrid: e.g., median for old movies where features are sparse, predictive for recent ones).

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor

# ----------------------------
# 1. Median-Year Imputer
# ----------------------------
def impute_budget_median(df):
    df_copy = df.copy()
    year_medians = df_copy.groupby("Year")["budget"].median()

    df_copy["budget_imputed"] = df_copy.apply(
        lambda row: row["budget"] 
        if not pd.isna(row["budget"]) 
        else year_medians.get(row["Year"], df_copy["budget"].median()), 
        axis=1
    )
    return df_copy


# ----------------------------
# 2. Predictive RF Imputer
# ----------------------------
def impute_budget_rf(df):
    df_copy = df.copy()
    features = ["Duration_clean", "Rating", "Votes_clean", "grossWorldWide"]

    # Training set: rows with budget + complete features
    train_data = df_copy.dropna(subset=["budget"] + features)
    if len(train_data) == 0:
        df_copy["budget_imputed"] = df_copy["budget"]
        return df_copy

    X_train, y_train = train_data[features], train_data["budget"]

    rf = RandomForestRegressor(
        n_estimators=200, max_depth=12, random_state=42, n_jobs=-1
    )
    rf.fit(X_train, y_train)

    df_copy["budget_imputed"] = df_copy["budget"]

    # Test set: rows with missing budget but complete features
    test_data = df_copy[df_copy["budget"].isna() & df_copy[features].notna().all(axis=1)]
    if len(test_data) > 0:
        df_copy.loc[test_data.index, "budget_imputed"] = rf.predict(test_data[features])

    return df_copy


# ----------------------------
# 3. Hybrid Imputer (RF + Median fallback)
# ----------------------------
def impute_budget_hybrid(df):
    df_copy = df.copy()
    year_medians = df_copy.groupby("Year")["budget"].median()
    features = ["Duration_clean", "Rating", "Votes_clean", "grossWorldWide"]

    # Train on complete cases
    train_data = df_copy.dropna(subset=["budget"] + features)
    if len(train_data) == 0:
        df_copy["budget_imputed"] = df_copy["budget"]
        return df_copy

    X_train, y_train = train_data[features], train_data["budget"]
    rf = RandomForestRegressor(
        n_estimators=200, max_depth=12, random_state=42, n_jobs=-1
    )
    rf.fit(X_train, y_train)

    df_copy["budget_imputed"] = df_copy["budget"]

    for idx, row in df_copy[df_copy["budget"].isna()].iterrows():
        # Default: year median
        pred = year_medians.get(row["Year"], df_copy["budget"].median())

        # Try RF prediction if features complete
        if not row[features].isna().any():
            pred_rf = rf.predict(row[features].values.reshape(1, -1))[0]
            if not np.isnan(pred_rf) and pred_rf > 0:
                pred = pred_rf

        df_copy.at[idx, "budget_imputed"] = pred

    return df_copy


# ----------------------------
# Evaluation Function
# ----------------------------
def evaluate_imputer(df, imputer_func, mask_fraction=0.2, random_state=42):
    df_copy = df.copy()

    # Only use rows where budget is known
    known = df_copy.dropna(subset=["budget"]).copy()

    # Randomly mask some budgets
    np.random.seed(random_state)
    mask_idx = np.random.choice(
        known.index, size=int(len(known) * mask_fraction), replace=False
    )
    known.loc[mask_idx, "budget"] = np.nan  # artificially remove

    # Apply imputer
    imputed_df = imputer_func(known)

    # Compare only masked rows
    y_true = df.loc[mask_idx, "budget"]
    y_pred = imputed_df.loc[mask_idx, "budget_imputed"]

    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred)
    return mae, rmse


# ----------------------------
# Run Evaluation
# ----------------------------
mae_median, rmse_median = evaluate_imputer(df, impute_budget_median)
mae_rf, rmse_rf = evaluate_imputer(df, impute_budget_rf)
mae_hybrid, rmse_hybrid = evaluate_imputer(df, impute_budget_hybrid)

print("Median-Year Imputation -> MAE:", mae_median, " RMSE:", rmse_median)
print("Predictive RF Imputation -> MAE:", mae_rf, " RMSE:", rmse_rf)
print("Hybrid (RF + Median-Year) -> MAE:", mae_hybrid, " RMSE:", rmse_hybrid)
```
